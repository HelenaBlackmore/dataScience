```python
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams

import numpy as np
from numpy.random import randn
```

```python
address='CARS.csv'
cars=pd.read_csv(address)
```

```python
# linechart
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

```python
mpg=(cars['MPG_City'] + cars['MPG_Highway'])/2

plt.plot(mpg)
```

```python
df=cars[['EngineSize', 'Horsepower', 'Weight']]
plt.plot(df)
```

```python
plt.bar(x,y)
```
