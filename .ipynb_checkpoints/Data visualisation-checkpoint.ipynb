### Labels and annotations: Formatting a graph


```python
#Functional method

# labels axes

x=range(1,10)
y=[1,2,3,4,.5,4,3,2,1]

plt.bar(x,y)
plt.xlabel('your x axis label')
plt.ylabel('y label')
```

```python
z=[1,2,3,4,.5]

veh_type=['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

```python
cars.head()

mpg=cars.MPG_City

fig=plt.figure()
ax=fig.add_axes([0.1,0.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
#ax.set_xticklabels(cars.make, rotation=60, fontsize='medium')  # this does not work , not sure why
ax.set_title('miles per galons in a city')

ax.set_xlabel('car make')
ax.set_ylabel('miles per galonn (city)')
```

```python
##legend

plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
#ax.set_xticklabels(cars.make, rotation=60, fontsize='medium')  # this does not work , not sure why
ax.set_title('miles per galons in a city')

ax.set_xlabel('car make')
ax.set_ylabel('miles per galonn (city)')
ax.legend(loc='best')
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,1,1])

mpg.plot()

ax.set_xticks(range(100))
#ax.set_xticklabels(cars.make, rotation=60, fontsize='medium')  # this does not work , not sure why
ax.set_title('miles per galons in a city')

ax.set_xlabel('car make')
ax.set_ylabel('miles per galonn (city)')
ax.legend(loc='best')
ax.set_ylim([0, 80])

ax.annotate('Tioyota Corolla', xy= (150, 60), xytext=(21,70), arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
import seaborn as sns

sns.set_style('whitegrid')
```

```python
#cars.head()
cars.index=cars.Make
mpg=cars['MPG_City']

mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sns.distplot(mpg)
```

```python
cars.plot(kind='scatter', x='Horsepower', y='MPG_City', c=['darkgrey'], s=10) # s= size of the marker, c= color
```

```python
# seaborn

sns.regplot(x='Horsepower', y='MPG_City', data=cars,scatter=True)
```
